## 오픈데이터 API 사용하기

1. API 신청하기 
 - 해당 데이터 상세에서 `오픈API` -> `활용신청` -> 내역입력

2. 참고자료에 `평택고등학교 주문형강좌 수업자료 활용` 내용 작성

3. 동의합니다 하고 활용신청 클릭

4. 마이페이지 -> OpenAPI -> 활용신청 현황 -> 활용신청한 데이터 클릭

5. 상세보기에서 일반 인증키(Decoding) 복사 후 인증키 설정 클릭

6. ApiKeyAuth2 (apiKey) 의 Value에 인증키 입력 이후 설정

7. API 목록에서 GET 부분 클릭 후 입력하고자 하는 설정 값 (Parameters) 입력

8. OpenAPI 실행 준비 -> OpenAPI 호출
